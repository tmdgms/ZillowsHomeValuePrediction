### 카테고리데이터의 상관관계 파악

```python
# heatingorsystemtypeid ~ airconditioningtypeid
```

```python
# pivot table로 변환

test3 = pd.DataFrame(m_data)
test3["count"] = 1

test4 = m_data.pivot_table(values="count", index=["heatingorsystemtypeid"], columns=["airconditioningtypeid"], aggfunc=sum)
test4
```

```python
# 확률 값 으로 변환

test44 = test4/test4.sum().sum()
test44
```

```python
# 상관 관계 파악을 위한 히트맵

sns.heatmap(test44,cmap=mpl.cm.bone_r, annot=True)
```

```python
# 베이지안 정리를 이용한 독립 여부확인

test44.loc[1]/test44.loc[1].sum(), test44.loc[1]
```

```python
test44.iloc[:, 0]/test44.iloc[:, 0].sum(), test44.iloc[:, 0]
```
