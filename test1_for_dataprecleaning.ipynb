```python
pro_2017 = pd.read_csv("C:\\dss\\team_project\\first\data\\properties_2017.csv")
pro_2017.tail()
```

```python
train_2017 = pd.read_csv("C:\\dss\\team_project\\first\data\\train_2017.csv")
train_2017.tail()
```

```python
test = pd.merge(pro_2017, train_2017)
result = test.sort_values(by=["transactiondate"]).reset_index(drop=True)
result
```

```python
# 결측치를 처리하기위한 방법 
# 1. 컬럼의 카테고리가 1개 밖에 없으면서 NaN값도 많고 logerror 분포도 안정적이지 않으면 그 컬럼은 삭제한다.
# 2. 0으로 채워넣기 = 갯수가 0인 것을 nan으로 처리하는 경우에 0을 넣어준다. (무엇인가의 갯수를 뜻하는 컬럼에서 0이 없으면 그부분은 0)
# result.fillna(result)
# 3. 평균으로 채워넣기 = 그냥 측정 못한 nan은 그 컬럼의 평균으로 채워넣는다.
# result.fillna(result.mean()[column])
```

```python
# 칼럼별 결측치 갯수
result.isnull().sum()
```

```python
plt.scatter(result["bathroomcnt"], result["logerror"])
plt.show()
```

```python
result_x = result[result.columns[0:58]]
result_y = result[result.columns[58]]
print(result_x, result_y)
```

```python
# logerror 에 대한 시간에 따른 그래프
plt.figure(figsize=(15,10))
plt.plot(result["logerror"])
plt.show()
```

```python
# log error에 대한 평균과 분산(표준편차)
mu = sum(result["logerror"])/len(result)
Var = sum((result["logerror"] - sum(result["logerror"])/len(result))*(result["logerror"] - sum(result["logerror"])/len(result)))/len(result["logerror"])
std = Var**0.5
print(mu, Var, std)
```

```python
# 원래 모델(집값 예측모델)의 오차값 = 잔차^2의 합
print(sum(result["logerror"]*result["logerror"])) 
```
